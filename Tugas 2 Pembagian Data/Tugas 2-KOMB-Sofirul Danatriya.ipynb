```python
# import library yang dibutuhkan
import pandas as pd

# memuat data file
df = pd.read_csv('iris.csv')

# preview data
df.head()
```

Dari total 150 dataset, 70% nya akan digunakan untuk training dan 30% nya untuk testing. Namun, pada file csv dataset diurutkan berdasarkan species(klasifikasi) sehingga diperlukan untuk merandom/mengacak data terlebih dahulu.

```python
# merandom baris pada dataset
random_df = df.sample(frac=1).reset_index(drop=True)

# membagi menjadi data training dan data testing
split_index = int( len(df) * 0.7)
train_df = random_df.iloc[:split_index,:]
test_df = random_df.iloc[split_index:,:]

# preview data training
train_df.head()
```

Setelah data dibagi menjadi 2. Maka data training akan dibagi menjadi 5 fold.

```python
folds = [] # array untuk menyimpan semua fold
k_fold = 5 # banyak fold

# membagi menjadi 5 fold
split_index = [x for x in range(0, len(train_df)+1, int(len(train_df)/k_fold))]
folds = [train_df.iloc[split_index[i]:split_index[i+1],:] for i in range(k_fold)]

# preview masing masing fold
for i in range(k_fold):
    print(f'Fold {i+1}')
    print(folds[i].head(2), '\n')
```

Data training sudah siap digunakan untuk membuat model 5 cross validation. Selanjutnya saya akan mendemokan contoh penggunaannya.

```python
# contoh penggunaan 5 cross validation
def createModel(folds):
    n = len(folds)
    for i in range(n):
        print("Model ",i+1," berlatih dengan :")
        for j in range(5):
            if j != i:
                print("fold-",(j%5)+1,folds[j].head(2), '\n')
        print("Model ",i+1," divalidasi dengan :")
        print("fold-",i+1,folds[i].head(2), '\n')

        print("="*50)


createModel(folds)
```
